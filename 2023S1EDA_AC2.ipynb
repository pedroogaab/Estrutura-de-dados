Escreva uma função chamada **compara** que receba como parâmetro uma pilha e uma fila. Ela deve retornar True caso as duas estruturas passadas como parâmetro sejam iguais e False caso contrário. Um pilha e uma fila são consideradas iguais tanto quando os elementos estão na mesma ordem ou em ordem contrária. Por exemplo, a pilha A=[0,1,2,3,4] é igual as filas B = [0,1,2,3,4] e C = [4,3,2,1,0].

 - Você deve usar as classes Pilha e Fila ensinadas na disciplina e disponibilizadas juntamente com este notebook. 

- As únicas operações permitidas para acessar os valores da pilha e fila são, respectivamente, as operações pop e dequeue. 

- Você deve garantir que ao final da função, a pilha e fila passadas como parâmetro estejam intactas, isto é, como os mesmos valores passados como entrada. 

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class Node:
    
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next


class Pilha:
    
    def __init__(self):
        self.head = None
        self.size = 0

    def size(self):
        return self.size


    def isEmpty(self):
        return self.head is None
    
    
    def push(self, data):
        
        new_noh = Node(data)
        
        if self.head is None:
            self.head = new_noh
        else: 
            new_noh.next = self.head
            self.head = new_noh
        self.size += 1


    def pop(self):
        if self.isEmpty():
            print("Pilha vazia")
            return None
        
        data = self.head.data
        self.head = self.head.next
        
        self.size -= 1
        return data 
    
    
    def getTop(self):

        
        if self.head is None:
            print("Pilha vazia")
            return None
        return self.head.data
    
    
class Fila:
    
    def __init__(self):
        self.front = None
        self.rear = None
        self.size = 0
        
        
    def size(self):
        return self.size
        
        
    def isEmpty(self):
        return self.front is None
    
    
    def enqueue(self, data):
        new_noh = Node(data)
        
        if self.rear is None:
            self.rear = new_noh
            self.front = self.rear
            
        else:
            self.rear.next = new_noh
            self.rear = new_noh
        
        self.size += 1
        
        
        
    def dequeue(self):
        
        if self.size == 0:
            print("Fila vazia")
            return None
        
        data = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None  
            
        self.size -= 1
        return data


def compara(pilha, fila):
    # Cria uma pilha temporária para guardar os valores e fazer as comparações
    if pilha.size != fila.size:
        return False
    
    pilha_temp = Pilha()
    pilha_temp.head = pilha.head
    pilha_temp.size = pilha.size
    

    # Cria uma fila temporária para guardar os valores e fazer as comparações
    fila_temp = Fila()
    fila_temp.front = fila.front
    fila_temp.rear = fila.rear
    fila_temp.size = fila.size
    
    
    nova_pilha = Pilha()

    inversa = Fila()
    reversa = Fila()
    
    
    acertos_inversa = 0
    acertos_reversa = 0
    
    num_interacoes = Fila()


    while not pilha_temp.isEmpty():
        valor_pilha = pilha_temp.pop()
        nova_pilha.push(valor_pilha)
        inversa.enqueue(valor_pilha)
        
    while not nova_pilha.isEmpty():
        valor_pilha = nova_pilha.pop()
        reversa.enqueue(valor_pilha)
        
        
        
    while not inversa.isEmpty() and not reversa.isEmpty() and not fila_temp.isEmpty():
        valor1 = inversa.dequeue()
        valor2 = reversa.dequeue()
        valor_principal = fila_temp.dequeue()
        
        num_interacoes.enqueue(valor_principal)
        
        if valor1 == valor_principal:
            acertos_inversa += 1
        if valor2 == valor_principal:
            acertos_reversa += 1
        
    
    if acertos_reversa == num_interacoes.size:
        return True
    elif acertos_inversa == num_interacoes.size:
        return True
    else: return False        
    

##########################################################################

# testando a função criada
pilha1 = Pilha()
pilha1.push(0)
pilha1.push(1)
pilha1.push(2)
pilha1.push(3)
pilha1.push(4)

# testando a função criada
pilha2 = Pilha()
pilha2.push(4)
pilha2.push(3)
pilha2.push(2)
pilha2.push(1)
pilha2.push(0)

fila1 = Fila()
fila1.enqueue(0)
fila1.enqueue(1)
fila1.enqueue(2)
fila1.enqueue(3)
fila1.enqueue(4)

fila2 = Fila()
fila2.enqueue(4)
fila2.enqueue(3)
fila2.enqueue(2)
fila2.enqueue(1)
fila2.enqueue(0)

fila3 = Fila()
fila3.enqueue(4)
fila3.enqueue(3)
fila3.enqueue(2)
fila3.enqueue(0)
fila3.enqueue(1)

comp1 = compara(pilha1, fila1)
comp2 = compara(pilha1, fila2)
comp3 = compara(pilha2, fila1)
comp4 = compara(pilha2, fila2)
comp5 = compara(pilha2, fila3)

print('Resultado retornado:')
print('\tComparacao 1?: ', comp1)
print('\tComparacao 2?: ', comp2)
print('\tComparacao 3?: ', comp3)
print('\tComparacao 4?: ', comp4)
print('\tComparacao 5?: ', comp5)

print('\n' + 20*'-')
print('Resultado esperado:')
print('\tComparacao 1?: True')
print('\tComparacao 2?: True')
print('\tComparacao 3?: True')
print('\tComparacao 4?: True')
print('\tComparacao 5?: False')
````

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class ListaInsertion:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def getSize(self):
        return self.size

    def isEmpty(self):
        return self.size == 0

    def __len__(self):
        return self.size

    def insert(self, data):
        newNode = Node(data)
        if self.isEmpty():
            self.head = newNode
            self.tail = newNode
        else:
            current = self.head
            while current and current.data <= data:
                current = current.next
            
            if not current: # inserção no final
                newNode.prev = self.tail
                self.tail.next = newNode
                self.tail = newNode
            elif not current.prev: # inserção no começo
                newNode.next = self.head
                self.head.prev = newNode
                self.head = newNode
            else: # inserção no meio
                newNode.prev = current.prev
                current.prev.next = newNode
                newNode.next = current
                current.prev = newNode
            
        self.size += 1

    def sublista(self, a, b):
        if a < 0 or b >= self.size or a > b:
            return None

        result = ListaInsertion()
        current = self.head
        i = 0

        while i < a:
            current = current.next
            i += 1

        while i <= b:
            result.insert(current.data)
            current = current.next
            i += 1

        return result

    def comparar(self, str1, str2):
        avg_chars1 = sum(map(len, str1.split())) / len(str1.split())
        avg_chars2 = sum(map(len, str2.split())) / len(str2.split())
        return int(avg_chars1) == int(avg_chars2)

    def repeticoes(self, n):
        return self._repeticoes(self.head, self.tail, n)

    def _repeticoes(self, first, last, n):
        if first == None or last == None or first.data == None or last.data == None:
            return 0
        
        if first == last:
            return self.comparar(first.data, n)
        
        mid = self._get_mid(first, last)
        
        if self.comparar(mid.data, n):
            return 1 + self._repeticoes(first, mid.prev, n) + self._repeticoes(mid.next, last, n)
        else:
            return self._repeticoes(first, mid.prev, n) + self._repeticoes(mid.next, last, n)
            
    def _get_mid(self, first, last):
        slow = first
        fast = first
        
        while fast != last and fast.next != last:
            slow = slow.next
            fast = fast.next.next
        
        return slow



##########################################################################

# testa o exercicio
lista = ListaInsertion()        

# inserir no inicio da lista
lista.insert("Centro Universitario Facens")
lista.insert("Eu fui para um shopping em Sorocaba")
lista.insert("A menina gosta de jogar")
lista.insert("O menino gosta de andar")
lista.insert("Tá tá tá")

print("\nLista:\n", lista)

qtd1 = lista.repeticoes(3)
qtd2 = lista.repeticoes(2)
qtd3 = lista.repeticoes(30)
print("Qtd. repeticoes 1: ", qtd1)
print("Qtd. repeticoes 2: ", qtd2)
print("Qtd. repeticoes 3: ", qtd3)

print('\n' + 20*'-')
print('Resultado esperado:')
print("\nLista:\n")
print('[Tá tá tá, A menina gosta de jogar, ', end="")
print('O menino gosta de andar, ', end="")
print('Eu fui para um shopping em Sorocaba, ', end="")
print('Centro Universitario Facens]', end="")
````

## Exercício 5

Crie uma classe chamada **PilhaFila** que implemente uma estrutura de dados que seja ao mesmo tempo uma pilha e fila dinâmica usando um deque dinâmico instanciado a partir da classe fornecida junto com esse notebook. Você deve implementar todos os métodos de uma fila e de um pilha dinâmica nessa estrutura aproveitando tudo o que for possível do deque. 

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

class pilhaFila:
    def __init__(self):
        
        self.get_top = None
        self.get_tail = None
        
        self.get_front = None
        self.get_rear = None
        
        self.get_size = 0
        
        self.pilha = False
        self.fila = False
        
    def size(self):
        return self.get_size
    
    
    def isEmpty(self):
        return self.get_size == 0


    def push(self, data):
        noh = Node(data)

        if self.get_top is None:
            self.get_top = noh
            self.get_tail = noh
        else:
            noh.next = self.get_top
            self.get_top = noh
            
        self.pilha = True
        self.fila = False
        self.get_size += 1
        
        
    def pop(self):
        if self.get_top is None:
            return self.isEmpty()

        data = self.get_top.data
        self.get_top = self.get_top.next
        self.get_size -= 1

        if self.get_top is None:
            self.get_tail = None

        self.pilha = True
        self.fila = False
        
        self.get_rear = self.get_top if self.get_top is not None else None
        return data






    def enqueue(self, data):
        noh = Node(data)

        if self.get_front is None:
            self.get_front = noh
            self.get_rear = noh
        else:
            self.get_rear.next = noh
            self.get_rear = noh
            
        self.pilha = False
        self.fila = True
        self.get_size += 1

    def dequeue(self):
        if self.get_front is None:
            return self.isEmpty()

        data = self.get_front.data
        self.get_front = self.get_front.next
        self.get_size -= 1

        if self.get_front is None:
            self.get_tail = None

        self.pilha = False
        self.fila = True
        self.get_rear = self.get_tail if self.get_tail is not None else None
        return data
        

    def top(self):
        if self.get_top is None:
            return self.isEmpty()

        return self.get_top.data


    def front(self):
        if self.get_front is None:
            return self.isEmpty()
        
        return self.get_front.data
        
        
    def rear(self):
        if self.get_rear is None:
            return self.isEmpty()
        else:
            return self.get_rear.data

                
    def __str__(self):
        _str = '['

        if self.fila:
            noh_atual = self.get_front
            while noh_atual is not None:
                _str += str(noh_atual.data)
                if noh_atual.next is not None:
                    _str += ', '
                noh_atual = noh_atual.next

        elif self.pilha:
            noh_atual = self.get_top
            while noh_atual is not None:
                _str += str(noh_atual.data)
                if noh_atual.next is not None:
                    _str += ', '
                noh_atual = noh_atual.next

        _str += ']'
        return _str



    def clear(self):

        while not self.isEmpty():
            data = self.pop()




##########################################################################
       
# testando a PilhaFila
pilhaFila1 = pilhaFila()

pilhaFila1.push(5)
print(pilhaFila1)

pilhaFila1.enqueue(3)
print(pilhaFila1)

pilhaFila1.push(9)
print(pilhaFila1)

print( 'Size: ', pilhaFila1.size() )

valor = pilhaFila1.pop()
print(pilhaFila1)

valor = pilhaFila1.dequeue()
print(pilhaFila1)

print( 'Vazia?: ', pilhaFila1.isEmpty() )

valor = pilhaFila1.pop()
print(pilhaFila1)

print( 'Vazia?: ', pilhaFila1.isEmpty() )

valor = pilhaFila1.pop()
print(pilhaFila1)

pilhaFila1.push(2)
print(pilhaFila1)

pilhaFila1.push(14)
print(pilhaFila1)

print(pilhaFila1.front())
print(pilhaFila1.rear())
print(pilhaFila1.top())

print( 'Vazia?: ', pilhaFila1.isEmpty() )

pilhaFila1.clear()
print( 'Vazia?: ', pilhaFila1.isEmpty() )

print('\n' + 20*'-')
print('Resultado esperado:')
print("\nLista:\n")
print('[ 5 ]')
print('[ 5, 3 ]')
print('[ 5, 3, 9 ]')
print('Size:  3')
print('[ 5, 3 ]')
print('[ 3 ]')
print('Vazia?:  False')
print('[ ]')
print('Vazia?:  True')
print('pilhaFila1 vazia')
print('[ ]')
print('[ 2 ]')
print('[ 2 14 ]')
print('2')
print('14')
print('14')
print('Vazia?:  False')
print('Vazia?:  True)')
````
