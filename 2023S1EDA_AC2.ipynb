## Exercício 1 

Escreva uma função chamada **compara** que receba como parâmetro uma pilha e uma fila. Ela deve retornar True caso as duas estruturas passadas como parâmetro sejam iguais e False caso contrário. Um pilha e uma fila são consideradas iguais tanto quando os elementos estão na mesma ordem ou em ordem contrária. Por exemplo, a pilha A=[0,1,2,3,4] é igual as filas B = [0,1,2,3,4] e C = [4,3,2,1,0].

 - Você deve usar as classes Pilha e Fila ensinadas na disciplina e disponibilizadas juntamente com este notebook. 

- As únicas operações permitidas para acessar os valores da pilha e fila são, respectivamente, as operações pop e dequeue. 

- Você deve garantir que ao final da função, a pilha e fila passadas como parâmetro estejam intactas, isto é, como os mesmos valores passados como entrada. 

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next


class Pilha:
    def __init__(self):
        self.head = None
        self.size = 0

    def size_pilha(self):
        return self.size

    def isEmpty(self):
        return self.head is None

    def push(self, data):
        new_noh = Node(data)

        if self.head is None:
            self.head = new_noh
        else:
            new_noh.next = self.head
            self.head = new_noh
        self.size += 1

    def pop(self):
        if self.isEmpty():
            print("Pilha vazia")
            return None

        data = self.head.data
        self.head = self.head.next

        self.size -= 1
        return data

    def getTop(self):
        if self.head is None:
            print("Pilha vazia")
            return None
        return self.head.data

    def __str__(self):
        auxPilha = Pilha()

        while not self.isEmpty():
            data = self.pop()
            auxPilha.push(data)

        strPilha = "["
        while not auxPilha.isEmpty():
            top = auxPilha.pop()

            self.push(top)

            strPilha += " " + str(top)

        strPilha += " ]"

        return strPilha


class Fila:
    def __init__(self):
        self.front = None
        self.rear = None
        self.size = 0

    def size_fila(self):
        return self.size

    def isEmpty(self):
        return self.front is None

    def enqueue(self, data):
        new_noh = Node(data)

        if self.rear is None:
            self.rear = new_noh
            self.front = self.rear

        else:
            self.rear.next = new_noh
            self.rear = new_noh

        self.size += 1

    def dequeue(self):
        if self.size == 0:
            print("Fila vazia")
            return None

        data = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None

        self.size -= 1
        return data

    def __str__(self):
        if self.isEmpty():
            return "Fila vazia"

        current = self.front
        string = ""
        while current:
            string += str(current) + " -> "
            current = current.next

        return string + "None"

    def __str__(self):
        """
        Retorna uma string com as informacoes da fila
        quando o objeto e chamado dentro de um print()
        ou dentro de um str().
        """

        # cria uma nova fila auxiliar
        auxfila = Fila()

        # retira os elementos da fila antiga e preenche a auxiliar
        for i in range(self.size):
            auxfila.enqueue(self.dequeue())

        # cria uma string com os elementos da fila auxiliar e
        # enquanto isso, preenche novamente a fila original
        strfila = "["
        for i in range(auxfila.size):
            front = auxfila.dequeue()

            self.enqueue(front)

            strfila += " " + str(front)

        strfila += " ]"

        return strfila


def compara(pilha, fila):
    # Cria uma pilha temporária para guardar os valores e fazer as comparações
    if pilha.size_pilha() != fila.size_fila():
        return False

    pilha_temp = Pilha()
    pilha_temp.head = pilha.head
    pilha_temp.size = pilha.size

    # Cria uma fila temporária para guardar os valores e fazer as comparações
    fila_temp = Fila()
    fila_temp.front = fila.front
    fila_temp.rear = fila.rear
    fila_temp.size = fila.size

    nova_pilha = Pilha()

    inversa = Fila()
    reversa = Fila()

    acertos_inversa = 0
    acertos_reversa = 0

    num_interacoes = Fila()

    while not pilha_temp.isEmpty():
        valor_pilha = pilha_temp.pop()
        nova_pilha.push(valor_pilha)
        inversa.enqueue(valor_pilha)

    while not nova_pilha.isEmpty():
        valor_pilha = nova_pilha.pop()
        reversa.enqueue(valor_pilha)

    while not inversa.isEmpty() and not reversa.isEmpty() and not fila_temp.isEmpty():
        valor1 = inversa.dequeue()
        valor2 = reversa.dequeue()
        valor_principal = fila_temp.dequeue()

        num_interacoes.enqueue(valor_principal)

        if valor1 == valor_principal:
            acertos_inversa += 1
        if valor2 == valor_principal:
            acertos_reversa += 1

    if acertos_reversa == num_interacoes.size_fila():
        return True
    elif acertos_inversa == num_interacoes.size_fila():
        return True
    else:
        return False


##########################################################################

# testando a função criada
pilha1 = Pilha()
pilha1.push(0)
pilha1.push(1)
pilha1.push(2)
pilha1.push(3)
pilha1.push(4)

# testando a função criada
pilha2 = Pilha()
pilha2.push(4)
pilha2.push(3)
pilha2.push(2)
pilha2.push(1)
pilha2.push(0)

fila1 = Fila()
fila1.enqueue(0)
fila1.enqueue(1)
fila1.enqueue(2)
fila1.enqueue(3)
fila1.enqueue(4)

fila2 = Fila()
fila2.enqueue(4)
fila2.enqueue(3)
fila2.enqueue(2)
fila2.enqueue(1)
fila2.enqueue(0)

fila3 = Fila()
fila3.enqueue(4)
fila3.enqueue(3)
fila3.enqueue(2)
fila3.enqueue(0)
fila3.enqueue(1)

comp1 = compara(pilha1, fila1)
comp2 = compara(pilha1, fila2)
comp3 = compara(pilha2, fila1)
comp4 = compara(pilha2, fila2)
comp5 = compara(pilha2, fila3)

print("Resultado retornado:")
print("\tComparacao 1?: ", comp1)
print("\tComparacao 2?: ", comp2)
print("\tComparacao 3?: ", comp3)
print("\tComparacao 4?: ", comp4)
print("\tComparacao 5?: ", comp5)

print("\n" + 20 * "-")
print("Resultado esperado:")
print("\tComparacao 1?: True")
print("\tComparacao 2?: True")
print("\tComparacao 3?: True")
print("\tComparacao 4?: True")
print("\tComparacao 5?: False")
````

Faça uma classe chamada **ListaNSE**. Essa classe deve implementar uma **lista não sequencial e estática**. Essa lista deve ter o mesmo comportamento de uma **lista não sequencial e dinâmica**, mas deve ser guardada dentro de uma lista em Python com capacidade limitada.

Essa classe deve possui os seguintes métodos:

- inserir(x, i): insere x na posição i da estrutura
- remover(i): remove o item que está na posição i da estrutura e retorna seu valor 
- buscar(x): buscar o item x na lista e retorna sua posição
- size(): retorna o número de elementos da lista
- clear(): remove os elementos da lista
- \_\_str\_\_(): imprime a estrutura

- Você deve usar a lista do Python apenas para inicializar a estrutura. Depois disso, nenhum método do Python pode ser usado, como append, insert, etc.

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class ListaNSE:
    def __init__(self, capacity):
        self.capacity = capacity
        self.tamanho = 0
        self.lista = [None] * capacity

    def insert(self, x, i):
        if i < -self.capacity or i >= self.capacity:
            return f"Index fora de range permitido -> [-{self.capacity}] até [{self.capacity - 1}]"
        if self.lista[i] is None:
            self.lista[i] = x
        else:
            new_index = i + 1
            while new_index < self.capacity and self.lista[new_index] is not None:
                new_index += 1
            if new_index >= self.capacity:
                return "Lista cheia"
            while new_index > i:
                self.lista[new_index] = self.lista[new_index - 1]
                new_index -= 1
            self.lista[i] = x
        self.tamanho += 1

    def remove(self, i):
        if i >= self.capacity or i < -self.capacity:
            return f"Index fora de range permitido -> [-{self.capacity}] até [{self.capacity - 1}]"
        elif i >= 0:
            self.lista[i] = None
        elif i < 0:
            self.lista[-i - 1] = None
        if self.tamanho == 0:
            return "Erro, não é possível retirar um valor de uma lista vazia"
        else:
            self.tamanho -= 1

    def buscar(self, x):
        for pos, b in enumerate(self.lista):
            if b == x:
                return pos

    def size(self):
        return self.tamanho

    def __str__(self):
        nova_lista = [None] * self.size()
        i = 0
        for x in self.lista:
            if x is not None:
                nova_lista[i] = x
                i += 1
        return str(nova_lista)

    def clear(self):
        for i in range(self.capacity):
            self.lista[i] = None
        self.tamanho = 0


##########################################################################

# testa a Lista
lista = ListaNSE(5)

# inserir no inicio da lista
lista.insert(5, 0)
print("insert(5,0): ", lista)

lista.insert(10, 0)
print("insert(10,0): ", lista)

lista.insert(24, 2)
print("insert(24,2): ", lista)

lista.insert(25, 1)
print("insert(25,1): ", lista)

lista.remove(2)
print("remove(2): ", lista)

print("Size: ", lista.size())

lista.clear()
print("Clear: ", lista)

print("\n" + 20 * "-")
print("Resultado esperado:")
print("insertHead(5,0): [5]")
print("insertHead(10,0): [10,5]")
print("insert(24,2): [10,5,24]")
print("insert(25,1): [10,25,5,24]")
print("remove(2): [10,25,24]")
print("Size: 3")
print("Clear: []")
````

Crie uma classe chamada **PilhaFila** que implemente uma estrutura de dados que seja ao mesmo tempo uma pilha e fila dinâmica usando um deque dinâmico instanciado a partir da classe fornecida junto com esse notebook.
````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next


class pilhaFila:
    def __init__(self):
        self.get_front = None
        self.get_rear = None
        self.get_size = 0

    def size(self):
        return self.get_size

    def isEmpty(self):
        return self.get_size == 0

    def push(self, data):
        noh = Node(data)

        if self.get_rear is None:
            self.get_rear = noh
            self.get_front = noh
        else:
            self.get_rear.next = noh
            self.get_rear = noh

        self.get_size += 1

    def pop(self):
        if self.isEmpty():
            print("pilhaFila vazia")
            return None

        del_noh = self.get_rear
        data = del_noh.data

        if self.get_size == 1:
            self.get_rear = None
            self.get_front = None

        else:
            noh_aux = self.get_front
            while noh_aux.next != del_noh:
                noh_aux = noh_aux.next

            self.get_rear = noh_aux
            noh_aux.next = None

        self.get_size -= 1
        del del_noh
        return data

    def enqueue(self, data):
        self.push(data)

    def dequeue(self):
        if self.get_front is None:
            return self.isEmpty()

        del_noh = self.get_front
        data = del_noh.data
        self.get_front = del_noh.next

        del del_noh
        self.get_size -= 1
        return data

    def top(self):
        if self.get_rear is None:
            print("pilhaFila vazia")
            return None

        return self.get_rear.data

    def front(self):
        if self.get_front is None:
            print("pilhaFila vazia")
            return None

        return self.get_front.data

    def rear(self):
        if self.get_rear is None:
            return self.isEmpty()
        else:
            return self.get_rear.data

    def __str__(self):
        node = self.get_front
        result_str = "["

        while node is not None:
            result_str += " " + str(node.data)
            node = node.next

            if node is not None:
                result_str += ","

        result_str += " ]"
        return result_str

    def clear(self):
        while not self.isEmpty():
            self.pop()


##########################################################################

# testando a PilhaFila
pilhaFila1 = pilhaFila()

pilhaFila1.push(5)
print(pilhaFila1)

pilhaFila1.enqueue(3)
print(pilhaFila1)

pilhaFila1.push(9)
print(pilhaFila1)

print("Size: ", pilhaFila1.size())

valor = pilhaFila1.pop()
print(pilhaFila1)

valor = pilhaFila1.dequeue()
print(pilhaFila1)

print("Vazia?: ", pilhaFila1.isEmpty())

valor = pilhaFila1.pop()
print(pilhaFila1)

print("Vazia?: ", pilhaFila1.isEmpty())

valor = pilhaFila1.pop()
print(pilhaFila1)

pilhaFila1.push(2)
print(pilhaFila1)

pilhaFila1.push(14)
print(pilhaFila1)

print(pilhaFila1.front())
print(pilhaFila1.rear())
print(pilhaFila1.top())

print("Vazia?: ", pilhaFila1.isEmpty())

pilhaFila1.clear()
print("Vazia?: ", pilhaFila1.isEmpty())

print("\n" + 20 * "-")
print("Resultado esperado:")
print("\nLista:\n")
print("[ 5 ]")
print("[ 5, 3 ]")
print("[ 5, 3, 9 ]")
print("Size:  3")
print("[ 5, 3 ]")
print("[ 3 ]")
print("Vazia?:  False")
print("[ ]")
print("Vazia?:  True")
print("pilhaFila1 vazia")
print("[ ]")
print("[ 2 ]")
print("[ 2 14 ]")
print("2")
print("14")
print("14")
print("Vazia?:  False")
print("Vazia?:  True)")
````

Modifique a árvore binária de busca ensinada nesta disciplina e dê o nome de **TreePath**. Acrescente nessa árvore um método chamado **caminhos**. Esse método deve preencher um objeto da lista duplamente ligada fornecida junto com este notebook com a soma do valores dos nós que estão no caminho de cada um dos nós folhas até a raiz da árvore. Por exemplo, considere a árvore apresentada abaixo:

```
            ___________61___________
           |                        |
       ____34_____              ____67____
      |           |            |          |    
      19__        50__      __66        __94__
          |           |    |           |      |  
         20           55   64          70     95
```

Se a árvore acima fosse apresentada como entrada, o método **caminhos** deveria retornar uma lista duplamente ligada com os seguintes valores: [134, 200, 258, 292, 317], que é formada pelas seguites somas: [61+34+19+20, 61+34+50+55, 61+67+66+64, 61+67+94+70, 61+67+94+95].

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################


class NodeLista:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next  # apontador para o noh da frente
        self.prev = prev  # apontador para o noh de tras


class Lista:
    """
    Lista duplamente ligada
    """

    def __init__(self, data=None):
        self.head = None
        self.tail = None
        self.size = 0

    def insertHead(self, data):

        tempNodeLista = NodeLista(data)

        if self.size == 0:
            self.head = tempNodeLista
            self.tail = tempNodeLista

        else:

            tempNodeLista.next = self.head
            self.head.prev = tempNodeLista
            self.head = tempNodeLista

        self.size += 1

    def insertTail(self, data):
        tempNodeLista = NodeLista(data)

        if self.size == 0:
            self.tail = tempNodeLista
            self.head = self.tail

        else:
            tempNodeLista.prev = self.tail

            self.tail.next = tempNodeLista

            self.tail = tempNodeLista

        self.size += 1

        if self.size > 0:

            tempNodeLista = self.head
            auxValor = tempNodeLista.data

            if auxValor == data:
                return 0

            # percorre o resto da lista ate que o apontador next do noh seja nulo
            # o que indica que chegou no tail
            idx = 1
            while tempNodeLista.next is not None:

                # vai para o proximo noh
                tempNodeLista = tempNodeLista.next

                auxValor = tempNodeLista.data

                if auxValor == data:
                    return idx

                idx += 1

        return -1

    def __str__(self):

        strfila = ""

        if self.size > 0:

            tempNodeLista = self.head
            valor = tempNodeLista.data

            strfila += " " + str(valor)

            # percorre o resto da lista ate que o apontador next do noh seja nulo
            # o que indica que chegou no tail
            while tempNodeLista.next is not None:

                # vai para o proximo noh
                tempNodeLista = tempNodeLista.next

                valor = tempNodeLista.data

                strfila += " " + str(valor)

        return "[" + strfila + "]"


class Node:
    def __init__(self, data):

        self.data = data
        self.right = None
        self.left = None


class TreePath:
    def __init__(self):
        self.root = None

    def insert(self, data, node=None):

        if self.root == None:
            self.root = Node(data)

        else:

            if node is None:
                node = self.root

            # se o valor do novo noh for menor que o valor do noh atual, insere na esquerda
            if data < node.data:

                if node.left is None:

                    node.left = Node(data)

                else:
                    self.insert(data, node.left)

            # se o valor do novo noh for maior ou igual que o valor do noh atual, insere na direita
            else:

                if node.right is None:
                    node.right = Node(data)

                else:
                    self.insert(data, node.right)

    def caminhos(self):

        resultado = Lista()
        soma = ""

        if self.root:
            self.caminhos_recursivo(self.root, [], resultado)

        aux = resultado.head

        while aux:
            if aux.next == None:
                soma += str(aux.data)
            else:
                soma += str(aux.data) + "," + " "

            aux = aux.next

        return "[" + soma + "]"

    def caminhos_recursivo(self, node, caminho, resultado):

        caminho.append(node.data)

        if not node.left and not node.right:

            somaCaminho = sum(caminho)
            resultado.insertTail(somaCaminho)

        # realizando a recursiva para percorrer os valores da árvore
        if node.left:

            self.caminhos_recursivo(node.left, caminho, resultado)

        if node.right:

            self.caminhos_recursivo(node.right, caminho, resultado)

        caminho.pop()


##########################################################################

tree1 = TreePath()
tree1.insert(61)
tree1.insert(34)
tree1.insert(67)
tree1.insert(19)
tree1.insert(50)
tree1.insert(66)
tree1.insert(94)
tree1.insert(20)
tree1.insert(55)
tree1.insert(64)
tree1.insert(70)
tree1.insert(95)

lista = tree1.caminhos()
print(lista)

print("\n" + 20 * "-")
print("Resultado esperado:")
print("[134, 200, 258, 292, 317]")
````
