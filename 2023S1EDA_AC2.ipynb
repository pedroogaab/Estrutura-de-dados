## Exercício 1 

Escreva uma função chamada **compara** que receba como parâmetro uma pilha e uma fila. Ela deve retornar True caso as duas estruturas passadas como parâmetro sejam iguais e False caso contrário. Um pilha e uma fila são consideradas iguais tanto quando os elementos estão na mesma ordem ou em ordem contrária. Por exemplo, a pilha A=[0,1,2,3,4] é igual as filas B = [0,1,2,3,4] e C = [4,3,2,1,0].

 - Você deve usar as classes Pilha e Fila ensinadas na disciplina e disponibilizadas juntamente com este notebook. 

- As únicas operações permitidas para acessar os valores da pilha e fila são, respectivamente, as operações pop e dequeue. 

- Você deve garantir que ao final da função, a pilha e fila passadas como parâmetro estejam intactas, isto é, como os mesmos valores passados como entrada. 

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

class Node:
    
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next


class Pilha:
    
    def __init__(self):
        self.head = None
        self.size = 0

    def size_pilha(self):
        return self.size


    def isEmpty(self):
        return self.head is None
    
    
    def push(self, data):
        
        new_noh = Node(data)
        
        if self.head is None:
            self.head = new_noh
        else: 
            new_noh.next = self.head
            self.head = new_noh
        self.size += 1


    def pop(self):
        if self.isEmpty():
            print("Pilha vazia")
            return None
        
        data = self.head.data
        self.head = self.head.next
        
        self.size -= 1
        return data 
    


class Fila:
    
    def __init__(self):
        self.front = None
        self.rear = None
        self.size = 0
        
        
    def size_fila(self):
        return self.size
        
        
    def isEmpty(self):
        return self.front is None
    
    
    def enqueue(self, data):
        new_noh = Node(data)
        
        if self.rear is None:
            self.rear = new_noh
            self.front = self.rear
            
        else:
            self.rear.next = new_noh
            self.rear = new_noh
        
        self.size += 1
        
        
    def dequeue(self):
        
        if self.size == 0:
            print("Fila vazia")
            return None
        
        data = self.front.data
        self.front = self.front.next
        if self.front is None:
            self.rear = None  
            
        self.size -= 1
        return data
    
    
    
def compara(pilha, fila):
    # Cria uma lista de pilha temporária para guardar os valores e fazer as comparações
    pilha_temp = []
    pilha_secundaria = Pilha()
    pilha_secundaria.head = pilha.head
    pilha_secundaria.size = pilha.size

    # Cria uma lista de fila temporária para guardar os valores e fazer as comparações
    fila_temp = []
    fila_secundaria = Fila()
    fila_secundaria.front = fila.front
    fila_secundaria.rear = fila.rear
    fila_secundaria.size = fila.size

    # Executa o looping enquanto as estruturas nao estiverem vazias
    while not pilha_secundaria.isEmpty() and not fila_secundaria.isEmpty():
        # Remove o valor do topo da pilha e o adiciona à lista de pilha temporária
        pilha_value = pilha_secundaria.pop()
        pilha_temp.append(pilha_value)

        # Remove o valor da frente da fila e o adiciona à lista de fila temporária
        fila_value = fila_secundaria.dequeue()
        fila_temp.append(fila_value)

    # Verifica se ambas as listas são iguais
    if pilha_temp == fila_temp:
        return True
    
    # Exibe a lista de trás pra frente e retorna True caso as listas sejam iguais
    elif pilha_temp == fila_temp[::-1]: 
        return True
    
    # Caso as listas não sejam iguais, retorna False
    else:
        return False



##########################################################################

# testando a função criada
pilha1 = Pilha()
pilha1.push(0)
pilha1.push(1)
pilha1.push(2)
pilha1.push(3)
pilha1.push(4)

pilha2 = Pilha()
pilha2.push(4)
pilha2.push(3)
pilha2.push(2)
pilha2.push(1)
pilha2.push(0)

fila1 = Fila()
fila1.enqueue(0)
fila1.enqueue(1)
fila1.enqueue(2)
fila1.enqueue(3)
fila1.enqueue(4)

fila2 = Fila()
fila2.enqueue(4)
fila2.enqueue(3)
fila2.enqueue(2)
fila2.enqueue(1)
fila2.enqueue(0)

fila3 = Fila()
fila3.enqueue(4)
fila3.enqueue(3)
fila3.enqueue(2)
fila3.enqueue(0)
fila3.enqueue(1)


comp1 = compara(pilha1, fila1)
comp2 = compara(pilha1, fila2)
comp3 = compara(pilha2, fila1)
comp4 = compara(pilha2, fila2)
comp5 = compara(pilha2, fila3)



print('Resultado retornado:')
print('\tComparacao 1?: ', comp1)
print('\tComparacao 2?: ', comp2)
print('\tComparacao 3?: ', comp3)
print('\tComparacao 4?: ', comp4)
print('\tComparacao 5?: ', comp5)

print('\n' + 20*'-')
print('Resultado esperado:')
print('\tComparacao 1?: True')
print('\tComparacao 2?: True')
print('\tComparacao 3?: True')
print('\tComparacao 4?: True')
print('\tComparacao 5?: False')
````

## Exercício 2 

Faça uma classe chamada **ListaNSE**. Essa classe deve implementar uma **lista não sequencial e estática**. Essa lista deve ter o mesmo comportamento de uma **lista não sequencial e dinâmica**, mas deve ser guardada dentro de uma lista em Python com capacidade limitada.

Essa classe deve possui os seguintes métodos:

- inserir(x, i): insere x na posição i da estrutura
- remover(i): remove o item que está na posição i da estrutura e retorna seu valor 
- buscar(x): buscar o item x na lista e retorna sua posição
- size(): retorna o número de elementos da lista
- clear(): remove os elementos da lista
- \_\_str\_\_(): imprime a estrutura

- Você deve usar a lista do Python apenas para inicializar a estrutura. Depois disso, nenhum método do Python pode ser usado, como append, insert, etc.

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################
class ListaNSE:
    # Guarda o valor da capacidade informada
    def __init__(self, capacity):
        self.capacity = capacity

        self.tamanho = 0
        # Cria uma lista do tamanho da capacidade informada e armazena o valor None em cada uma
        self.lista = [None]*capacity
        
        
    def insert(self, x, i):
        if i <= self.capacity*(-1) or i >= self.capacity:
            return (f"Index fora de range permitido -> [-{self.capacity}] até [{self.capacity}]") 
        
        if self.lista[i] is None:
            self.lista[i] = x
            
        else:
            new_index = i + 1
            while new_index < self.capacity and self.lista[new_index] is not None:
                new_index += 1
                
            if new_index >= self.capacity:
                return "Lista cheia"
            
            while new_index > i:
                self.lista[new_index] = self.lista[new_index-1]
                new_index -= 1
                
            self.lista[i] = x
            
        self.tamanho += 1

        
  
        
    def remove(self, i):
        
        if i > self.capacity or i < self.capacity*(-1):
            return (f"Index fora de range permitido -> [-{self.capacity}] até [{self.capacity}]") 
        
        elif i >= 0 and i <= self.capacity:
            self.lista[i] = None
        
        elif i < 0:
            self.lista[-i] = None
            
        if self.tamanho == 0:
            return ("Erro, nao é possivel retirar um valor de uma lista vazia")
        else: self.tamanho -= 1
            
        
    
    def buscar(self, x):
        for pos, b in enumerate(self.lista):
            if b == x:
                return pos

    
    def size(self):
        return self.tamanho
    
    
    def __str__(self):
        nova_lista = [None]*self.size()
        i = 0
        for x in self.lista:
            if x is not None:
                nova_lista[i] = x
                i += 1
        return str(nova_lista)


    def clear(self):
        for i in range(self.capacity):
            self.lista[i] = None
        self.tamanho = 0

##########################################################################

# testa a Lista
lista = ListaNSE(5)        

# inserir no inicio da lista
lista.insert(5,0)
print( 'insert(5,0): ', lista )

lista.insert(10,0)
print( 'insert(10,0): ', lista )

lista.insert(24,2)
print( 'insert(24,2): ', lista )

lista.insert(25,1)
print( 'insert(25,1): ', lista )

lista.remove(2)
print( 'remove(2): ', lista )

print( 'Size: ', lista.size() )

lista.clear()
print( 'Clear: ', lista )

print('\n' + 20*'-')
print('Resultado esperado:')
print( 'insertHead(5,0): [5]')
print( 'insertHead(10,0): [10,5]')
print( 'insert(24,2): [10,5,24]')
print( 'insert(25,1): [10,25,5,24]')
print( 'remove(2): [10,25,24]')
print( 'Size: 3')
print( 'Clear: []')
````
